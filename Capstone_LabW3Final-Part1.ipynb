```python
import numpy as np # library to handle array data
import pandas as pd # library for data analsysis
```

```python
toronto_df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
toronto_df
```

**Dropping "Not assigned" in Borough**

```python
toronto_dropNotAssigned = toronto_df[toronto_df.Borough != 'Not assigned'].reset_index(drop=True)
toronto_dropNotAssigned
```

**Grouping different neighborhoods and separate them with a comma**

```python
toronto_grouped = toronto_dropNotAssigned.groupby(['Postal Code','Borough'], as_index=False).agg(lambda x: ','.join(x))
toronto_grouped
```

**Assign an empty neighborhoods with Borough values**

```python
mask = toronto_grouped['Neighborhood'] == "Not assigned"
toronto_grouped.loc[mask, 'Neighborhood'] = toronto_grouped.loc[mask, 'Borough']
toronto_grouped
```

**Print number of rows and columns**

```python
toronto = toronto_grouped.shape
print("The data has Rows {}, Columns {}".format(*toronto))
```
